## 籍贯分布地图

```python
from pyecharts import Page

from pyecharts import Geo
import pandas as pd 

page = Page()
df=pd.read_excel('info.xlsx')
df = df.round(3)
geo_jiguan_coords={df.iloc[i]['name']:[df.iloc[i]['籍贯经度'],df.iloc[i]['籍贯纬度']] for i in range(len(df))}
attr=list(df['name'])
df["地区"] = 0
for i in range(len(df)):
    if float(df.iloc[i]['籍贯经度']) >= 103 and float(df.iloc[i]['籍贯经度']) < 135 and float(df.iloc[i]['籍贯纬度']) >= 42 and float(df.iloc[i]['籍贯纬度']) < 53:
        df.loc[i,'地区'] = 1
    elif float(df.iloc[i]['籍贯经度']) < 103 and float(df.iloc[i]['籍贯经度']) > 73 and float(df.iloc[i]['籍贯纬度']) >= 18 and float(df.iloc[i]['籍贯纬度']) < 53:
        df.loc[i,'地区'] = 2
    elif float(df.iloc[i]['籍贯经度']) >= 103 and float(df.iloc[i]['籍贯经度']) < 135 and float(df.iloc[i]['籍贯纬度']) >= 32 and float(df.iloc[i]['籍贯纬度']) < 42:
        df.loc[i,'地区'] = 3
    elif float(df.iloc[i]['籍贯经度']) >= 103 and float(df.iloc[i]['籍贯经度']) < 135 and float(df.iloc[i]['籍贯纬度']) >= 18 and float(df.iloc[i]['籍贯纬度']) < 32:
        df.loc[i,'地区'] = 4
    elif float(df.iloc[i]['籍贯经度']) >= 135 or float(df.iloc[i]['籍贯经度']) < 73 or float(df.iloc[i]['籍贯纬度']) < 18 or float(df.iloc[i]['籍贯纬度']) > 53:
        df.loc[i,'地区'] = 5
    else:
        df.loc[i,'地区'] = 0
value = list(df['地区'])
jiguan = Geo("长江学者籍贯经纬度", title_color="#fff",
          title_pos="center", width=1200,
          height=600, background_color='#404a59')
jiguan.add("", attr,value, visual_range=[0, 6], visual_text_color="#fff",
        symbol_size=15, is_visualmap=True,is_piecewise=True, visual_split_number=5,geo_cities_coords=geo_jiguan_coords)
page.add(jiguan)
```
